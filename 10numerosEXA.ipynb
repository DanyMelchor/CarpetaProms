```python
AR=[]
lar=9
for i in range(0,10):
  AR.append(int(input("Ingresa un numero cualquiera: ")))
print(" ")
print("--arreglo segun el usuario--")
print(AR)
for i in range(1,10):
  for h in range(0,10-i):
    if(AR[h]>AR[h+1]):
      var=AR[h]
      AR[h]=AR[h+1]
      AR[h+1]=var
print(" ")
print("--arreglo de menos a mayor--")
print(AR)
for i in range(0,5):
  Var=AR[lar]
  AR[lar]=AR[i]
  AR[i]=Var
  lar=lar-1
print(" ")
print("--arreglo de mayor a menor--")
print(AR)
```
